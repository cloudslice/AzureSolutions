```python
%%pyspark 

from pyspark.sql.types import * 

from pyspark.sql.functions import * 

 

Customerschema = StructType([ 

    StructField("Registeration no", StringType()), 

    StructField("Quarter", IntegerType()), 

    StructField("Brand", StringType()), 

    StructField("Name", StringType()), 

    StructField("Address", StringType()), 

    StructField("Telephone_Number", StringType()), 

    ])
df = spark.read.load('Files/Sales/Customers/*.csv', format='csv', schema=Customerschema)
display(df.limit(100)) 
```

```python
df.printSchema() 
```

```python
##Namedf = df['Name', 'Brand'] 
Namedf = df.select("Name", "Address").where(df['Brand']=='honda') 

print(Namedf.count()) 

print(Namedf.distinct().count()) 

display(Namedf.distinct()) 
```
